# ggplot2: A data visualization package
### ggplot2 is designed with the grammar of graphics in mind hence the gg (grammer of graphics)

#### Three basic grammatical units
* Data - information you want visualized
* geometries - describes the shapes you want data in (dots, bars, line, etc.)
* aesthetics - visual attributes of the plot

```r
#Load libraries
library(readr)
library(dplyr)
library(ggplot2)
```

```r
#Load data
airbnb_data <- read_csv("/Users/luke/Downloads/AB_NYC_2019.csv")
```

```r
airbnb_data
```

```r
#create ggplot object
#add data
visual <- ggplot(airbnb_data)
```

```r
visual
```

```r
#adding aesthetics
graph1 <- ggplot(data = airbnb_data, aes(x = room_type, y = price))
```

```r
graph1
```

```r
#add geom to graph
graph1 <- ggplot(data = airbnb_data, aes(x = room_type, y = price)) + geom_point()
```

```r
graph1
```

```r
#add aesthetic to geom
graph1 <- ggplot(data = airbnb_data, aes(x = room_type, y = price)) + geom_point(aes(color = calculated_host_listings_count))
```

```r
graph1
```

### More Aesthetics:
* x
* y
* alpha (opacity)
* color
* fill
* group
* shape
* size
* stroke

```r
#manually changing Aesthetics
#making opacity of points .5
graph1 <- ggplot(data = airbnb_data, aes(x = room_type, y = price)) + 
geom_point(aes(color = calculated_host_listings_count), alpha = 0.5)
```

```r
graph1
```

### Add Labels
#### use labs() function to add labels

```r
graph1 + labs(title = "Airbnb NY", subtitle = "2019", y = "Price", x = "Room Type", color = "# of host listings")
```

### Saving visualizations as a local file
#### ggsave()

```r
ggsave("graph1-example.png")
```

## Example 1 (Bar Graph)

```r
bar_graph <- ggplot(data = airbnb_data, aes(neighbourhood_group)) + geom_bar(aes(fill=neighbourhood_group)) +
labs(title="Airbnb Data for NYC", subtitle="From Kaggle (2019)", x = "NY Borough", y = "# of Airbnbs")
```

```r
bar_graph
```

```r
ggsave("bar_graph.png")
```

## Example 2 (Scatter Plot, Fifa Data)

```r
#Only look at top 50 fifa athletes
fifa_top50 <- head(fifa_data,50)
fifa_top50
```
